```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Sample data
data = {
    'ReviewID': [1, 2, 3, 4, 5],
    'ReviewText': [
        "I love this product! It works great and exceeded my expectations.",
        "Terrible experience, the product broke within a week.",
        "Pretty good, but I had some issues with the setup.",
        "Amazing! I would buy it again.",
        "Not worth the money."
    ]
}


df = pd.DataFrame(data)


def preprocess_text(text):
    # Remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Convert to lowercase
    text = text.lower()
 
    tokens = word_tokenize(text)
  
    tokens = [word for word in tokens if word not in stopwords.words('english')]
  
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)

df['CleanedReviewText'] = df['ReviewText'].apply(preprocess_text)

print(df[['ReviewID', 'CleanedReviewText']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from collections import Counter
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')

# Sample data
data = {
    'ReviewID': [1, 2, 3, 4, 5],
    'ReviewText': [
        "I love this product! It works great and exceeded my expectations.",
        "Terrible experience, the product broke within a week.",
        "Pretty good, but I had some issues with the setup.",
        "Amazing! I would buy it again.",
        "Not worth the money."
    ]
}

df = pd.DataFrame(data)
```

```python

all_reviews = ' '.join(df['ReviewText'])

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color ='white').generate(all_reviews)

# Display word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Customer Reviews')
plt.show()
```

```python
def plot_ngrams(text, n=2, top_k=10):
    tokens = word_tokenize(text)
    ngrams = nltk.ngrams(tokens, n)
    ngram_freq = Counter(ngrams)
    common_ngrams = ngram_freq.most_common(top_k)
    
    ngrams_df = pd.DataFrame(common_ngrams, columns=['Ngram', 'Frequency'])
    ngrams_df['Ngram'] = ngrams_df['Ngram'].apply(lambda x: ' '.join(x))
    
    plt.figure(figsize=(10, 5))
    sns.barplot(x='Frequency', y='Ngram', data=ngrams_df)
    plt.title(f' Top {top_k} {n}-grams')
    plt.xlabel('Frequency')
    plt.ylabel(f'{n}-gram')
    plt.show()

plot_ngrams(all_reviews, n=2, top_k=10)

plot_ngrams(all_reviews, n=3, top_k=10)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()

# Compute sentiment scores
df['SentimentScore'] = df['ReviewText'].apply(lambda x: sid.polarity_scores(x)['compound'])

# Sentiment distribution plot
plt.figure(figsize=(10, 5))
sns.histplot(df['SentimentScore'], bins=20, kde=True)
plt.title('Sentiment Score Distribution')
plt.xlabel('Sentiment Score')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

# Fit and transform the text data
X_bow = vectorizer.fit_transform(df['ReviewText'])

df_bow = pd.DataFrame(X_bow.toarray(), columns=vectorizer.get_feature_names_out())
print(df_bow.head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the text data
X_tfidf = tfidf_vectorizer.fit_transform(df['ReviewText'])

df_tfidf = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print(df_tfidf.head())
```

```python
from gensim.models import Word2Vec
import numpy as np

tokenized_reviews = df['ReviewText'].apply(word_tokenize)

w2v_model = Word2Vec(sentences=tokenized_reviews, vector_size=100, window=5, min_count=1, workers=4)

# Generate average word vectors for each review
def get_avg_word_vector(tokens, model):
    vectors = [model.wv[word] for word in tokens if word in model.wv]
    if vectors:
        return np.mean(vectors, axis=0)
    else:
        return np.zeros(model.vector_size)

df['Word2Vec'] = tokenized_reviews.apply(lambda x: get_avg_word_vector(x, w2v_model))
print(df['Word2Vec'].head())
```
